# Dropping the Classification and Application Type columns, adding more neurons and 2 layers.

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
app_df_new = application_df.drop(columns=["EIN", "NAME", "CLASSIFICATION", "APPLICATION_TYPE"])
app_df_new
```

```python
y_new = app_df_new['IS_SUCCESSFUL']
X_new = app_df_new.drop(columns='IS_SUCCESSFUL')
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
X_new = pd.get_dummies(X_new)
X_new
```

```python
len(X_new.columns)
```

```python
# Split the preprocessed data into a training and testing dataset
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y_new, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler_new = scaler.fit(X_train_new)

# Scale the data
X_train_scaled_new = X_scaler_new.transform(X_train_new)
X_test_scaled_new = X_scaler_new.transform(X_test_new)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
 # Define the deep learning model 
nn_model_new = tf.keras.models.Sequential()

# First hidden layer
nn_model_new.add(tf.keras.layers.Dense(units=300, activation="relu", input_dim=28))

# Second hidden layer
nn_model_new.add(tf.keras.layers.Dense(units=100, activation="relu"))

# Third hidden layer
nn_model_new.add(tf.keras.layers.Dense(units=100, activation="relu"))

# Fourth hidden layer
nn_model_new.add(tf.keras.layers.Dense(units=100, activation="relu"))

# Output layer
nn_model_new.add(tf.keras.layers.Dense(units=1, activation="tanh"))

# Check the structure of the model
nn_model_new.summary()
```

```python
# Compile the model
nn_model_new.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model_new = nn_model_new.fit(X_train_scaled_new, y_train_new, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model_new.evaluate(X_test_scaled_new,y_test_new,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn_model_new.save('model.h5')
```
